## Problem: 6

Create an array $x$ with 100 points in the interval $[-1,1]$ (the points must be uniformly spaced). Then create another array $y$. 

Where $y = x^2 sin(x^2) + x$. 

Try to plot $y$ vs $x$ using matplotlib.

```python
# Write your code here
x = np.linspace(start = -1, stop = 1, num = 100)
print(x)
import matplotlib.pyplot as plt

y = x**2 * np.sin(x**2) + x

plt.figure(figsize = (10,7))
plt.plot(x, y, 'r')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Plot of y = x² sin(x²) + x")
plt.grid()

plt.show()
```

Create a random $3×5$ array using the `np.random.rand(3 ,5)` function and compute: the sum of all the entries, the sum of the rows, and the sum of the columns.

```python
# Write your code here

a = np.random.rand(3,5)
print(f"Random array = {a}")

s1 = np.sum(a)
print(f"Sum of all entries = {s1}")

s2 = np.sum(a, axis = 1)
print(f"Sum of all the rows = {s2}")

s3 = np.sum(a, axis = 0)
print(f"Sum of all the rows = {s3}")
```
